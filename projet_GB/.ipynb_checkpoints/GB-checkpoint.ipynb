```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
def tetha(a, b, t, gamma_dot=1.0):
    # Paramètres physiques selon Abkarian (Section 1.3.3)
    A = -0.5  # Vorticité (négatif pour rotation horaire) [cite: 901]
    B = 0.5 * (a**2 - b**2) / (a**2 + b**2) # Forme [cite: 902]
    rp=b/a
    
    # Période de Jeffery (tau) [cite: 800]
    tau = (2 * np.pi / gamma_dot) * (rp + 1/rp)
    
    # Coefficient effectif
    coeff = np.sqrt((A + B) / (A - B))
    
    # Calcul de l'angle périodique de base (entre -pi/2 et pi/2)
    # On utilise t/tau pour la progression temporelle
    theta_base = - np.arctan(coeff * np.tan(np.pi * t / (tau/2)))
    
    # Déroulement manuel pour une courbe continue
    # floor(t / (tau/2)) compte le nombre de demi-périodes passées
    n_half_turns = np.floor(t / (tau / 2) + 0.5)
    y = theta_base - np.pi * n_half_turns
    
    return  y

class FCN(nn.Module):
    
    def __init__(self, N_INPUT, N_OUTPUT, N_HIDDEN, N_LAYERS):
        super().__init__()
        activation = nn.Tanh
        self.fcs = nn.Sequential(
            nn.Linear(N_INPUT, N_HIDDEN),
            activation()
        )
        self.fch = nn.Sequential(*[
                        nn.Sequential(
                            nn.Linear(N_HIDDEN, N_HIDDEN),
                            activation()) for _ in range(N_LAYERS-1)])
        self.fce = nn.Linear(N_HIDDEN, N_OUTPUT)
        
        # Initialisation spécifique des poids (Xavier/Glorot) 
        self._init_weights()

    def _init_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Linear):
                # Xavier normal est optimal pour les activations Tanh
                nn.init.xavier_normal_(m.weight)
                nn.init.zeros_(m.bias)
        
    def forward(self, x):
        x = self.fcs(x)
        x = self.fch(x)
        x = self.fce(x)
        return x
```

```python
# condition initiale 
a ,b = 4,1.5
gamma_dot = 1.0
rp = b / a
# Période exacte de Jeffery / KS rigide
period = 0.25*(2 * np.pi / gamma_dot) * (rp + 1/rp) 
print(period)

# Utilisez au moins une période complète pour l'entraînement
t = torch.linspace(0, period, 500).view(-1,1)
y = tetha(a, b, t).view(-1,1)
print(t.shape, y.shape)
# simulation 
t_data = t[0:200:20]
y_data = y[0:200:20]
print(t_data.shape, y_data.shape)
print(y_data)

plt.figure()
plt.plot(t, y, label="Exact solution")
plt.scatter(t_data, y_data, color="tab:orange", label="Training data")
plt.legend()
plt.show()
```

```python
t_physics = torch.linspace(0, period, 500).view(-1, 1).requires_grad_(True)

# 3. Boucle d'entraînement
model = FCN(1, 1, 32, 3)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

for epoch in range(6001): # Augmente le nombre d'époques, les PINNs sont lents
    optimizer.zero_grad()
    
    # 1. Prédiction globale
    x_pred = model(t_physics) 
    # 2. Conditions Initiales (Extraites du premier point de t_physics)
    # t_physics[0] vaut 0.0
    x_0_pred = x_pred[0]
    loss_x0 = (x_0_pred)**2
    # 3 Prédiction sur les points de mesure (les 10 points orange)
    y_data_pred = model(t_data) 
    loss_data = torch.mean((y_data_pred - y_data)**2)
    
    # 5. Total Loss avec pondération dynamique
    total_loss = 500 * loss_x0 + 300*loss_data
    
    total_loss.backward()
    optimizer.step()

    if epoch % 1000 == 0:
        print(f"Epoch {epoch}: Loss {total_loss.item():.6f}")
```

```python
plt.figure(figsize=(10, 6))

# Solution exacte (Ligne bleue)
plt.plot(t_physics.detach(),tetha(a, b, t).detach(), label="Solution exacte", color="tab:red", alpha=0.7)

# Points d'entraînement (Points orange)
plt.scatter(t_data, y_data, color="tab:orange", label="Points d'entraînement (Data)", zorder=5)

# Prédiction PINN (Ligne rouge)
yh = model(t).detach()
y_pred = model(t_physics).detach()
plt.plot(t,yh, color="tab:blue", linewidth=4, alpha=0.8, label="Neural network prediction",linestyle="--")

plt.title("PINN : Apprentissage hybride (Physique + Data)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
a, b = 4.0, 1.5
gamma_dot = 1.0
A = -0.5
B = 0.5 * (a**2 - b**2) / (a**2 + b**2)
# Période théorique pour couvrir une rotation complète 
period = 0.25*(2 * np.pi) / (gamma_dot * np.sqrt(0.5**2 - B**2)) 

# Temps pour la physique (Couvre la période complète)
t_physics = torch.linspace(0, period, 500).view(-1, 1).requires_grad_(True)

# Modèle et Optimiseur
model = FCN(1, 1, 32, 3)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

for epoch in range(12001):
    optimizer.zero_grad()
    
    # 1. Prédiction sur les points de collocation (Physique)
    theta_pred = model(t_physics)
    
    # 2. Calcul de la dérivée exacte [Différenciation Automatique]
    dtheta_dt = torch.autograd.grad(
        theta_pred, t_physics, 
        torch.ones_like(theta_pred), 
        create_graph=True
    )[0]

    # 3. Résidu de l'EDO Keller-Skalak (Régime rigide) [cite: 901, 902]
    # L'équation est : dθ/dt = γ * (A + B*cos(2θ))
    physics_residu = dtheta_dt - gamma_dot * (A + B * torch.cos(2 * theta_pred))
    loss_physics = torch.mean(physics_residu**2)
    
    # 4. Condition Initiale (t=0 => theta=0)
    # On utilise theta_pred[0] car t_physics[0] est 0.0
    loss_x0 = theta_pred[0]**2 
    
    # 5. Perte sur les données expérimentales (Points orange)
    y_data_pred = model(t_data) 
    loss_data = torch.mean((y_data_pred - y_data)**2)
    w_phys = 0.5 if epoch < 3000 else 1.0
    w_data = 100.0
    w_ci   = 500.0
    
    total_loss = w_phys * loss_physics + w_ci * loss_x0 + w_data * loss_data
    
    total_loss.backward()
    optimizer.step()

    if epoch % 1000 == 0:
        print(f"Epoch {epoch}: Loss {total_loss.item():.6f} (Phys: {loss_physics.item():.4f})")
```

```python
plt.figure(figsize=(10, 6))

# Solution exacte (Ligne bleue)
plt.plot(t_physics.detach(),tetha(a, b, t).detach(), label="Solution exacte", color="tab:red", alpha=0.7)

# Points d'entraînement (Points orange)
plt.scatter(t_data, y_data, color="tab:orange", label="Points d'entraînement (Data)", zorder=5)

# Prédiction PINN (Ligne rouge)
yh = model(t).detach()
y_pred = model(t_physics).detach()
plt.plot(t,yh, color="tab:blue", linewidth=4, alpha=0.8, label="Neural network prediction",linestyle="--")

plt.title("PINN : Apprentissage hybride (Physique + Data)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
import os

# --- 1. Paramètres Physiques (Régime Rigide) ---
a, b = 4.0, 1.5
gamma_dot = 1.0
A = -0.5  # Vorticité
B = 0.5 * (a**2 - b**2) / (a**2 + b**2)  # Coefficient de forme [cite: 902]

# Calcul de la période de référence (tau) [cite: 800]
tau_ref = np.pi / (gamma_dot * np.sqrt(0.5**2 - B**2))
periods_to_test = [2 * tau_ref, tau_ref, 0.5 * tau_ref]
labels = ["Période Complète", "Demi-Période", "Quart de Période"]

# --- 2. Fonction Solution Exacte (Déroulée/Continue) ---
def get_exact_solution(a, b, t, gamma_dot=1.0):
    # Paramètres physiques selon Abkarian 
    A = -0.5  # Vorticité (négatif pour rotation horaire)
    B = 0.5 * (a**2 - b**2) / (a**2 + b**2) # Forme
    rp=b/a
    tau = (2 * np.pi / gamma_dot) * (rp + 1/rp)
    coeff = np.sqrt((A + B) / (A - B))
    theta_base = - np.arctan(coeff * np.tan(np.pi * t / (tau/2)))
    n_half_turns = np.floor(t / (tau / 2) + 0.5)
    y = theta_base - np.pi * n_half_turns
    
    return  y

# --- 3. Boucle d'entraînement sur les 3 périodes ---
results = {}

if not os.path.exists('plots'): os.makedirs('plots')

for p_idx, period in enumerate(periods_to_test):
    print(f"\n--- Entraînement : {labels[p_idx]} (Durée: {period:.2f}s) ---")
    
    # Réinitialisation du modèle et de l'optimiseur
    torch.manual_seed(123)
    model = FCN(1, 1, 32, 3) 
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
    
    # Points de collocation pour la physique
    t_physics = torch.linspace(0, period, 500).view(-1, 1).requires_grad_(True)
    
    # Filtrage des données expérimentales (t_data, y_data supposés définis)
    mask = t_data.flatten() <= period
    t_train = t_data[mask].view(-1, 1)
    y_train = y_data[mask].view(-1, 1)

    loss_history = []

    for epoch in range(12001):
        optimizer.zero_grad()
        
        # A. Perte de données (Data Loss)
        yh = model(t_train)
        loss_data = torch.mean((yh - y_train)**2)
        
        # B. Perte physique (EDO Residual) [cite: 901]
        yhp = model(t_physics)
        # dθ/dt
        dtheta_dt = torch.autograd.grad(yhp, t_physics, torch.ones_like(yhp), create_graph=True)[0]
        
        # Résidu : dθ/dt - γ(A + B*cos(2θ)) = 0
        physics_res = dtheta_dt - gamma_dot * (A + B * torch.cos(2 * yhp))
        loss_physics = torch.mean(physics_res**2)
        
        # C. Perte Condition Initiale (t=0, θ=0)
        loss_ci = model(torch.tensor([[0.0]]))**2
        
        # Pondération hybride
        w_phys = 1.0 if epoch > 3000 else 0.1
        total_loss = 100 * loss_data + w_phys * loss_physics + 500 * loss_ci
        
        total_loss.backward()
        optimizer.step()
        
        loss_history.append(loss_physics.item())

        # Affichage et sauvegarde périodique
        if (epoch + 1) % 4000 == 0:
            print(f"Epoch {epoch+1}: Loss {total_loss.item():.6f} (Phys: {loss_physics.item():.4f})")

    # Sauvegarde des résultats pour comparaison finale
    results[labels[p_idx]] = {
        "t": t_physics.detach().numpy(),
        "pred": model(t_physics).detach().numpy(),
        "exact": get_exact_solution(a, b, t_physics.detach().numpy()),
        "loss": loss_history
    }
```

```python
plt.figure(figsize=(18, 5))

for i, (label, data) in enumerate(results.items()):
    plt.subplot(1, 3, i+1)
    
    # Tracé Solution Exacte (Rouge) et PINN (Bleu pointillé)
    plt.plot(data["t"], data["exact"], color="tab:red", label="Exacte (Continue)", alpha=0.7)
    plt.plot(data["t"], data["pred"], color="tab:blue", linestyle="--", linewidth=3, label="PINN Prediction")
    
    # Points de données originaux
    mask = t_data.flatten() <= data["t"].max()
    plt.scatter(t_data[mask], y_data[mask], color="tab:orange", s=30, label="Train Data", zorder=5)
    
    plt.title(f"{label}")
    plt.xlabel("Temps (s)")
    if i == 0: plt.ylabel("Angle θ (rad)")
    plt.legend()
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import torch
import torch.nn as nn
import numpy as np
import os

# --- 1. Paramètres Physiques (Régime Rigide) [cite: 901, 902] ---
a, b = 4.0, 1.5
gamma_dot = 1.0
A = -0.5  # Vorticité [cite: 901]
B = 0.5 * (a**2 - b**2) / (a**2 + b**2)  # Coefficient de forme 

# Calcul de la période de référence (tau) [cite: 800]
tau_ref = np.pi / (gamma_dot * np.sqrt(0.5**2 - B**2))
periods_to_test = [2 * tau_ref, tau_ref, 0.5 * tau_ref]
labels = ["Période Complète", "Demi-Période", "Quart de Période"]

# --- 2. Boucle d'entraînement sur les 3 périodes ---
results = {}

for p_idx, period in enumerate(periods_to_test):
    print(f"\n--- Entraînement : {labels[p_idx]} (Durée: {period:.2f}s) ---")
    
    torch.manual_seed(123)
    model = FCN(1, 1, 32, 3) 
    
    # Points de collocation pour la physique
    t_physics = torch.linspace(0, period, 500).view(-1, 1).requires_grad_(True)
    
    # Filtrage des données expérimentales
    mask = t_data.flatten() <= period
    t_train = t_data[mask].view(-1, 1)
    y_train = y_data[mask].view(-1, 1)

    # --- PHASE 1 : Pré-entraînement avec ADAM ---
    optimizer_adam = torch.optim.Adam(model.parameters(), lr=1e-3)
    for epoch in range(13001):
        
        optimizer_adam.zero_grad()
        
        yhp = model(t_physics)
        dtheta_dt = torch.autograd.grad(yhp, t_physics, torch.ones_like(yhp), create_graph=True)[0]
        
        # Résidu : dθ/dt - γ(A + B*cos(2θ)) = 0 [cite: 901]
        loss_physics = torch.mean((dtheta_dt - gamma_dot * (A + B * torch.cos(2 * yhp)))**2)
        loss_data = torch.mean((model(t_train) - y_train)**2)
        loss_ci = model(torch.tensor([[0.0]]))**2
        lambda_physics = 0.1 if epoch < 2000 else 20.0 
        
        total_loss = 100 * loss_data + lambda_physics * loss_physics + 500 * loss_ci
        total_loss.backward()
        optimizer_adam.step()

    # --- PHASE 2 : Affinement avec L-BFGS ---
    print("Passage à l'optimiseur L-BFGS...")
    optimizer_lbfgs = torch.optim.LBFGS(
        model.parameters(), 
        lr=1, 
        max_iter=5000, # Itérations internes par appel .step()
        line_search_fn="strong_wolfe" 
    )

    def closure():
        optimizer_lbfgs.zero_grad()
        yhp = model(t_physics)
        dtheta_dt = torch.autograd.grad(yhp, t_physics, torch.ones_like(yhp), create_graph=True)[0]
        
        loss_physics = torch.mean((dtheta_dt - gamma_dot * (A + B * torch.cos(2 * yhp)))**2)
        loss_data = torch.mean((model(t_train) - y_train)**2)
        loss_ci = model(torch.tensor([[0.0]]))**2
        
        total_loss = 100 * loss_data + 1.0 * loss_physics + 500 * loss_ci
        total_loss.backward()
        return total_loss

    optimizer_lbfgs.step(closure)

    # Sauvegarde des résultats
    results[labels[p_idx]] = {
        "t": t_physics.detach().numpy(),
        "pred": model(t_physics).detach().numpy(),
        "exact": get_exact_solution(a, b, t_physics.detach().numpy()),
    }
```

```python
plt.figure(figsize=(18, 5))

for i, (label, data) in enumerate(results.items()):
    plt.subplot(1, 3, i+1)
    
    # Tracé Solution Exacte (Rouge) et PINN (Bleu pointillé)
    plt.plot(data["t"], data["exact"], color="tab:red", label="Exacte (Continue)", alpha=0.7)
    plt.plot(data["t"], data["pred"], color="tab:blue", linestyle="--", linewidth=3, label="PINN Prediction")
    
    # Points de données originaux
    mask = t_data.flatten() <= data["t"].max()
    plt.scatter(t_data[mask], y_data[mask], color="tab:orange", s=30, label="Train Data", zorder=5)
    
    plt.title(f"{label}")
    plt.xlabel("Temps (s)")
    if i == 0: plt.ylabel("Angle θ (rad)")
    plt.legend()
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
